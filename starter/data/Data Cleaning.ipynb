```python
# remove all spaces
for col in df.columns:
    if df_cleanup[col].dtype != type(str):
        orig_type = df_cleanup[col].dtype
        df_cleanup[col] = df_cleanup[col].astype(str)
        df_cleanup[col] = df_cleanup[col].str.strip()
        df_cleanup[col] = df_cleanup[col].astype(orig_type)
    else:
        df_cleanup[col] = df_cleanup[col].str.strip()
```
